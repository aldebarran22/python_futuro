```python
dt.drop(columns=['idempleado', 'idempresa'], inplace=True)
```

```python
dt['iva'] = round(dt.importe * 0.21, 2)
```

```python
dt
```

```python
dt['total'] = dt.importe + dt.iva
```

```python
dt = dt[['idpedido','cliente','importe','iva','total','pais']]
dt
```

```python
dt2 = dt[(dt.pais=='Alemania') & (dt.importe >= 30) & (dt.importe <= 60)]
dt2
```

```python
dt2.to_excel('pedidos_alemania.xlsx', index=False, float_format='%.2f')
```

```python
paises = dt.pais.unique()
paises
```

```python
for nombrePais in paises:
    dtaux = dt[dt.pais==nombrePais]
    nombreFichero = f"paises/{nombrePais}.xlsx"
    dtaux.to_excel(nombreFichero, index=False, float_format='%.2f')
    print(f'exportando {nombrePais} con ',dtaux.shape[0],'filas')
```
