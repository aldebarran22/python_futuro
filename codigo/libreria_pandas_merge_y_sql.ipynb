## Pandas operaciones de merge

```python
import pandas as pd
import sqlite3 as dbapi
from os.path import isfile
import numpy as np
```

```python
dtP = pd.read_csv('Demo_Pandas/Pedidos.csv', sep=';')
print(dtP.shape, dtP.columns)

dtEm = pd.read_csv('Demo_Pandas/Empleados.txt', sep=';')
print(dtEm.shape, dtEm.columns)
```

```python
dtFinal = pd.merge(dtP, dtEm, left_on='idempleado',right_on='id')
dtFinal.head()
```

## Pandas acceso a base de datos

```python
path = 'Demo_Pandas/empresa3.db'
if isfile(path):
    con = dbapi.connect(path)
    sql = "select * from productos"
    
    dtSQL = pd.read_sql(sql, con)

else:
    print('No existe el fichero')
    
```

```python
dtSQL['precio'] = np.round(dtSQL.precio)
dtSQL.head()
```
