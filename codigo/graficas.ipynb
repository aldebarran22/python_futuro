- Segundo 

```python
import matplotlib.pyplot as plt
from random import randint
import numpy as np
```

```python
x1 = list(range(10))
y1 = [randint(10,30) for i in range(10)]
```

```python
x2 = list(range(10))
y2 = [randint(10,30) for i in range(10)]
```

```python
x3 = list(range(10))
y3 = [randint(10,30) for i in range(10)]
```

Generamos un gráfico

```python
# Comentario
plt.subplot2grid((2,2),(0,0))
plt.plot(x1,y1)

plt.subplot2grid((2,2),(0,1))
plt.plot(x2,y2)

plt.subplot2grid((2,2),(1,0), colspan=2)
plt.plot(x3,y3)
plt.title('Gráficas')
plt.show()
```

```python
x = list(range(20))

y1 = np.array([randint(10,30) for i in range(20)])
y2 = [randint(10,30) for i in range(20)]
y3 = [i**2 for i in range(20)]
y4 = [i**1/2 for i in range(20)]


plt.plot(x, y1, label="y1")
plt.plot(x, y2, label="y2")
plt.scatter(x, y3, label="y3",color="red")
plt.plot(x,y4, label="y4")
plt.grid(True)
plt.legend()
plt.xlabel('eje X')
plt.ylabel('eje Y')
plt.title("Gráficas")
plt.text( 7.5, 100,'Punto de corte')

plt.show()

```
