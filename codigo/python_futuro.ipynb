Separar los pedidos en un fichero para cada país. En formato csv, json, html

```python
paises = df.pais.unique()
for pais in paises:
    df2 = df[df.pais==pais]
    path = f"paises/{pais}.json"
    print(f'Generando {path}')
    df2.to_json(path, orient='records', indent=4)
```
